```python
#https://docs.python.org/3.4/library/xml.etree.elementtree.html
respTree = ET.fromstring(response.text)
```

```python
respTree.tag
```

```python
respTree.attrib
```

```python
for child in respTree:
    print(child.tag, child.attrib)
```

```python
for r in respTree.iter():
    print(r)
```

```python
import pandas as pd
contactdf = pd.DataFrame(columns=["ContactID", "LastName", "FirstName"])

for c in respTree.iter("Contact"):
    print(c.attrib["id"])
    print("Last Name:", c.find("LastName").text)
    try:
        print("First Name:", c.find("FirstName").text)
        fname = c.find("FirstName").text
    except:
        print("NoFirstName")
        fname = "NoFirstName"
    contactdf.append({'ContactID':c.attrib["id"], 'LastName':c.find("LastName").text, 'FirstName':fname}, ignore_index=True)
```
